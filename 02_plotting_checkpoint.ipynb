This notebook is used for checking 3plex data. <br>
More specifically, the singleplex and multiplex data are plotted side by side for each primer in each assay.

# Import libraries

```python
import pandas as pd
from pathlib import Path
```

```python
import python_libraries.plotting_function as plotfunc
import python_libraries.utilities as utils
```

```python
utils.colourblind()
NMETA = 6
```

# Load data

```python
curve_type_dict = {0: "raw_rb", 
                   1: "norm_curve",
                   2: "fitted_param",
                   3: "fitted_curve",
                   4: "norm_fitted_param",
                   5: "norm_fitted_curve",
                   6: 'c_param',
                   7: 'norm_c_param'}

# select the curve types based on the above dictionary information
selected_curve = 1
CURVE_TYPE = curve_type_dict[selected_curve]
```

```python
data_path = f'{Path.cwd().parent}/data'
data_selected_path = f'{data_path}/20210701_3_plex'
```

```python
singleplex_id = "20210701_01"
multiplex_id = "20210701_02"
```

```python
df_curve_single = pd.read_csv(f'{data_selected_path}/processed_data/{singleplex_id}_{CURVE_TYPE}.csv', index_col=0)
df_curve_single = utils.order_columns(df_curve_single, NMETA)
df_curve_multi = pd.read_csv(f'{data_selected_path}/processed_data/{multiplex_id}_{CURVE_TYPE}.csv', index_col=0)
df_curve_multi = utils.order_columns(df_curve_multi, NMETA)
```

# VISUALIZATION CHECK POINT

## Plotting filtered RAW curves

```python
print('SINGLEPLEX CURVES')
plotfunc.curve_plot_by(df_curve_single, NMETA, col_number=3)
```

```python
print('MULTIPLEX CURVES')
plotfunc.curve_plot_by(df_curve_multi, NMETA, col_number=3)
```

## plotting S2M curves

```python
plotfunc.plot_comparison_S2M(df_curve_single, df_curve_multi, NMETA)
```
